```python
import pyvisa

rm = pyvisa.ResourceManager()
print(rm.list_resources())

for address in rm.list_resources():
    try:
        inst = rm.open_resource(address)
        inst.timeout = 2000  # ms

        idn = inst.query('*IDN?').strip()
        print(f'{address}  -->  {idn}')

        inst.close()

    except Exception as e:
        print(f'{address}  -->  no response / not SCPI: {e}')
```

| Method    | Purpose                      |
| --------- | ---------------------------- |
| `write()` | send command                 |
| `read()`  | read response                |
| `query()` | send command + read response |
| `close()` | close connection             |


```python
scope = rm.open_resource('TCPIP0::k-dx3024g-60147.local::inst0::INSTR')
# query: send command, wait for response, return response as string
print(scope.query('*IDN?'))
# equivalent to:
scope.write("*IDN?")
response = scope.read()
print(response)


print(scope.query(":TIMebase:SCALe?"))
print(scope.query(":CHANnel1:SCALe?"))

scope.write(":WAVeform:SOURce CHANnel1")
scope.write(":WAVeform:FORMat ASCii")

data = scope.query(":WAVeform:DATA?")
print(data[:200])
```
